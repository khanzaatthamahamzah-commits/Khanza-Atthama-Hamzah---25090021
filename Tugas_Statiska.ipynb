```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Membaca dataset dari Google Drive
df = pd.read_csv('/content/drive/MyDrive/dataset/MovieData-2018-2020.csv')

# Menghapus data yang kosong pada genre dan budget
df = df.dropna(subset=['genre', 'budget'])

# Mengatasi genre yang jumlah datanya hanya 1
genre_count = df['genre'].value_counts()
df = df[df['genre'].isin(genre_count[genre_count >= 2].index)]


# Stratified Random Sampling berdasarkan genre
train, test = train_test_split(
    df,
    test_size=0.3,
    stratify=df['genre'],
    random_state=21
)

# Mengambil hanya 30 data teratas hasil sampling dari test set
sample = test.head(30)

# Menampilkan sample
print(sample[['name', 'genre', 'budget']])
```

```python
import scipy.stats as stats
import math

# Mengambil data budget dari sample
budget = sample['budget']

# Statistik sample
n = len(budget)
x_bar = budget.mean()
s = budget.std(ddof=1)

# Confidence Level 95%
confidence_level = 0.90
alpha = 1 - confidence_level
dfree = n - 1

# Nilai t kritis
t_critical = stats.t.ppf(1 - alpha/2, dfree)

# Margin of Error
margin_error = t_critical * (s / math.sqrt(n))

# Confidence Interval
lower = x_bar - margin_error
upper = x_bar + margin_error

print("Jumlah Sampel :", n)
print("Mean Budget :", x_bar)
print("Standar Deviasi :", s)
print("Margin of Error :", margin_error)
print(f"Confidence Interval 90% : ({lower}, {upper})")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Data budget dari 30 sampel
budget = sample['budget']

# Mean dan standar deviasi
mean = budget.mean()
std = budget.std(ddof=1)

# Membuat histogram
plt.figure(figsize=(10,6))
plt.hist(
    budget,
    bins=10,
    density=True,
    alpha=0.6,
    edgecolor='black'
)

# Membuat kurva distribusi (bentuk gunung)
x = np.linspace(budget.min(), budget.max(), 1000)
y = norm.pdf(x, mean, std)

plt.plot(x, y, linewidth=3)

# Menampilkan garis rata-rata
plt.axvline(mean, linestyle='--', linewidth=2,
            label=f'Mean = {mean:,.0f}')

plt.title('Distribusi Budget Film (30 Data Sample)')
plt.xlabel('Budget Film')
plt.ylabel('Density')
plt.legend()
plt.show()
```
